```python
def estimate_coefficients(size, price): 
    # size of the dataset OR number of observations/points 
    n = np.size(size)
  
    # mean of x and y
    # Since we are using numpy just calling mean on numpy is sufficient 
    mean_x, mean_y = np.mean(size), np.mean(price) 
  
    # calculating cross-deviation and deviation about x 
    SS_xy = np.sum(price*size - n*mean_y*mean_x) 
    SS_xx = np.sum(size*size - n*mean_x*mean_x) 
  
    # calculating regression coefficients 
    b_1 = SS_xy / SS_xx 
    b_0 = mean_y - b_1*mean_x 
  
    return(b_0, b_1)

    # x,y are the location of points on graph
    # color of the points change it to red blue orange play around



def plot_regression_line(size,price, b): 
    # plotting the points as per dataset on a graph
    plt.figure(figsize=(12,6))
    plt.scatter(size, price, color = "m",marker = "o", s = 30) 

    # predicted response vector 
    y_pred = b[0] + b[1]*size 
  
    # plotting the regression line
    plt.plot(size, y_pred, color = "g")
  
    # putting labels for x and y axis
    
    plt.xlabel('Size of flat in sqm', size=16) 
    plt.ylabel('Price', size=16) 
  
    # function to show plotted graph
    plt.show()
```

# 1. Linear relationship

<font size="10">A linear relationship between the independent and dependent variables</font>


```python
plt.style.use('seaborn-whitegrid')
from sklearn.datasets.samples_generator import make_regression
x1, y1 = make_regression(n_samples=100, n_features=1, noise=10)
plt.figure(figsize=(15,6))
plt.plot(x1, y1, 'o', color='black');
plt.title("Linear Relationship Exists", size=16)
```

```python
plt.style.use('seaborn-whitegrid')
from sklearn.datasets.samples_generator import make_regression
x1, y1 = make_regression(n_samples=100, n_features=1, noise=9000)
plt.figure(figsize=(18,8))
plt.plot(x1, y1, 'o', color='black');
plt.title("Non-linear Relationship Exists", size=16)
```

# Normal distribution

<font size="10"> The variable follows a normal distribution.</font>

<font size="10"> Q-Q Plot</font>

```python
import scipy.stats as stats
plt.figure(figsize=(15,4))
stats.probplot(x1[:,0], dist="norm", plot=plt)
plt.show()
```

The points must lie on this line to conclude that it follows a normal distribution. In our case, yes it does! 

# There is little or no Multicollinearity

```python
#create sample dataset with 3 x features
x3, y3 = make_regression(n_samples=100, n_features=3, noise=20)

#convert to a pandas dataframe
import pandas as pd
df = pd.DataFrame(x3)
df.columns = ['x1','x2','x3']

#generate correlation matrix
corr = df.corr()

corr.style.background_gradient(cmap='coolwarm')
```

If the correlation is >=0.8, the multicollinearity assumption is being broken.
